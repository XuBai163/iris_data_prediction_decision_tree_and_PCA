```python
# import required packages
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn.metrics import accuracy_score
from matplotlib import pyplot as plt
```

```python
# load data
data = pd.read_csv('iris_data.csv')
data.head()
```

```python
# define X and y
X = data.drop(['target', 'label'], axis = 1)
y = data.loc[:,'label']
```

```python
# set up decision tree model
dc_tree = tree.DecisionTreeClassifier(criterion = 'entropy', min_samples_leaf = 5)
```

```python
# train the model
dc_tree.fit(X,y)
```

```python
# evaluate the model 
y_predict = dc_tree.predict(X)
accuracy = accuracy_score(y, y_predict)
print(accuracy)
```

```python
# visualize the tree
fig1 = plt.figure(figsize = (10, 10))
tree.plot_tree(dc_tree, filled = 'True', feature_names = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth'], class_names = ['Setoas', 'Versicolor', 'Virginica'])
plt.show()
```

```python
# set up a decision tree model with min_sample_leaf = 1
dc_tree = tree.DecisionTreeClassifier(criterion = 'entropy', min_samples_leaf = 1)
dc_tree.fit(X,y)
fig1 = plt.figure(figsize = (15, 15))
tree.plot_tree(dc_tree, filled = 'True', feature_names = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth'], class_names = ['Setoas', 'Versicolor', 'Virginica'])
plt.show()
```
